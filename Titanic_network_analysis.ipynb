### 1. Graph Construction

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')
from bokeh.plotting import show

defaults = dict(width=1000, height=800, padding=0.1,
                xaxis=None, yaxis=None)
hv.opts.defaults(
    opts.EdgePaths(**defaults), opts.Graph(**defaults), opts.Nodes(**defaults))
```

```python
edges = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/NLP & NA/edges.csv')
nodes = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/NLP & NA/nodes.csv')
```

```python
edges
```

```python
nodes
```

```python
# mapping character names from Id
id_to_label = dict(zip(nodes['Id'], nodes['Label']))

edges['Source Name'] = edges['Source'].map(id_to_label)
edges['Target Name'] = edges['Target'].map(id_to_label)

edges_with_names = edges[['Source Name', 'Target Name', 'Weight']]
edges_with_names
```

```python
g = nx.from_pandas_edgelist(edges_with_names,
                            source='Source Name',
                            target='Target Name',
                            edge_attr='Weight',
                            create_using = nx.DiGraph)
```

```python
pos = nx.layout.kamada_kawai_layout(g)
```

```python
plt.figure(figsize=(10, 10))
nx.draw(g, pos, with_labels=True, node_color="skyblue", edge_color="gray", node_size=300, font_size=8)
plt.show()
```

```python
g_plot = hv.Graph.from_networkx(g, pos).opts(tools=['hover'])
show(hv.render(g_plot))
```

### 2. Graph Analysis
Determinate the basic graph indicators:
**Total number of nodes and edges**: to understand the *scale and complexity* of the graph
**Average degree**: giving an average of *connections per node*
**Graph density**:  providing an insight into the *overall connectivity*

```python
print(f"Graph info:")
print(f"Number of nodes: {g.number_of_nodes()}")
print(f"Number of edges: {g.number_of_edges()}")

avg_degree = sum(dict(g.degree()).values()) / g.number_of_nodes()
print(f"Average degree: {round(avg_degree, 3)}")

print(f"Graph density: {round(nx.density(g), 3)}")
print(f"Graph is directed: {g.is_directed()}")
```

**Graph Interpretation:**
1. Number of Nodes: There are 70 unique characters in the graph involved in the interactions.
2. Number of Edges: There are 299 connections or co-occurrences between characters. This number gives a sense of the interaction frequency and scene co-presence in the storyline.
3. Average Degree: On average, each character interacts with approximately 8.5 other characters, suggesting that key characters are involved in multiple interactions.
4. Graph Density: With a density of 0.062, the graph is fairly sparse, meaning that only about 6.2% of possible edges between characters exist. There are significant interactions among characters, however, each character is not interacting with all other characters. This aligns with the social dynamics of the movie structure, where characters from different classes do not interact much.

### 3. Node analysis
Calculate basic centralities and compare the most central characters.

```python
degree_cent = nx.degree_centrality(g)
between_cent = nx.betweenness_centrality(g, weight='weight')
eigen_cent = nx.eigenvector_centrality(g, weight='weight')

nx.set_node_attributes(g, degree_cent, 'degree_cent')
nx.set_node_attributes(g, between_cent, 'between_cent')
nx.set_node_attributes(g, eigen_cent, 'eigen_cent')
```

```python
# Degree Centrality
g_plot_degree = hv.Graph.from_networkx(g, pos).opts(
    tools=['hover'],
    node_size=hv.dim('degree_cent') * 100,
    node_color='degree_cent',
    cmap='Blues',
    title="Degree Centrality"
)
show(hv.render(g_plot_degree))
```

```python
# Betweenness Centrality
g_plot_betweenness = hv.Graph.from_networkx(g, pos).opts(
    tools=['hover'],
    node_size=hv.dim('between_cent') * 200,
    node_color='between_cent',
    cmap='Oranges',
    title="Betweenness Centrality"
)
show(hv.render(g_plot_betweenness))
```

```python
# Eigenvector Centrality
g_plot_eigenvector = hv.Graph.from_networkx(g, pos).opts(
    tools=['hover'],
    node_size=hv.dim('eigen_cent') * 150,
    node_color='eigen_cent',
    cmap='Purples',
    title="Eigenvector Centrality"
)
show(hv.render(g_plot_eigenvector))
```

```python
# Comparison Dataframe
centrality_df = pd.DataFrame({
    'Degree Centrality': degree_cent,
    'Betweenness Centrality': between_cent,
    'Eigenvector Centrality': eigen_cent
})
```

```python
# Top Characters by Degree Centrality
top_degree = centrality_df.sort_values(by='Degree Centrality', ascending=False).head(10)
print("Top 10 characters by Degree Centrality:")
top_degree
```

```python
# Top Characters by Betweenness Centrality
top_betweenness = centrality_df.sort_values(by='Betweenness Centrality', ascending=False).head(10)
print("Top 10 characters by Betweenness Centrality:")
top_betweenness
```

```python
# Top Characters by Eigenvector Centrality
top_eigenvector = centrality_df.sort_values(by='Eigenvector Centrality', ascending=False).head(10)
print("Top 10 characters by Eigenvector Centrality:")
top_eigenvector
```

**Comparison Between Centralities of Characters:**

Degree Centrality: It measures how connected a node is by counting its direct connections to other nodes.

Betweenness Centrality: It Indicates how often a node lies on the shortest path between other nodes, and how they connect different parts of the network acting as bridges.

Eigenvector Centrality: It considers influence of nodes that are are connected to other stronger nodes, giving more weight to nodes connected to other highly connected nodes.

* Rose is highly central by all metrics, with both high direct and bridge connectivity, highlighting her role in connecting various characters and also shows greater influence, as she is connected with other strong nodes or characters.

* Jack is highly connected and bridges groups, but his influence is slightly lower in terms of indirect connections and relationship with other strong nodes.

* Andrews, though less directly connected, holds high influence due to his connections with other influential characters, likely due to his role as the ship's designer.


### 4. Node attributes
Add Gender attribute (from dataset to the network) and calculate and/or visualize the share of females and males characters in the film. Furthermore, subgraph network to only male characters. Who are the main characters, what is their relationship?



```python
# Adding gender attribute to the graph nodes
gender_dict = dict(zip(nodes['Label'], nodes['Gender']))
nx.set_node_attributes(g, gender_dict, 'gender')

gender_counts = pd.Series(gender_dict).value_counts()
print("Gender distribution in the network:")
print(gender_counts)

# Gender distribution
gender_counts.plot(kind='bar', title='Gender Distribution in the Titanic Network', color=['skyblue', 'lightcoral'])
plt.show()
```

```python
# subgraph of only male characters
male_characters = [n for n, attr in g.nodes(data=True) if attr['gender'] == 'M']
male_subgraph = g.subgraph(male_characters)

# Visualize the male character subgraph

degree_centrality_male = nx.degree_centrality(male_subgraph)
nx.set_node_attributes(male_subgraph, degree_centrality_male, 'degree_centrality_male')

pos = nx.layout.kamada_kawai_layout(male_subgraph)
g_plot = hv.Graph.from_networkx(male_subgraph, pos).opts(
    edge_alpha=0.5,
    tools=['hover'],
    node_size=hv.dim('degree_centrality_male') * 150,
    node_color='degree_centrality_male',
    cmap='Blues',
    title="Male Character Subgraph"
)

show(hv.render(g_plot))
```

```python
male_centrality_df = pd.DataFrame({
    'Degree Centrality': degree_centrality_male,
    'Betweenness Centrality': nx.betweenness_centrality(male_subgraph,  weight='weight'),
    'Eigenvector Centrality': nx.eigenvector_centrality(male_subgraph,  weight='weight')
})

# Top 5 male characters
top_male_characters = male_centrality_df.sort_values(by='Degree Centrality', ascending=False).head(5)
print("Top 5 central male characters:")
top_male_characters
```
